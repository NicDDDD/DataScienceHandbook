Linear Regression Prep

```python
# import data and packages
import numpy as np
import matplotlib.pyplot as plt
import scipy as sc
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn import preprocessing 
data = pd.read_csv('LinReg.csv') #load data
%matplotlib inline
```

```python
#print(data.head()) 
y = data['output']
X = data.drop('output',axis=1).values #axis=1 indictaes colomns

#Making train and test data
m,n = X.shape

shuffIndx = np.random.permutation(np.arange(m))
# np.arange() is like linspace for matlab

X,Y = X[shuffIndx], y[shuffIndx]
X_train = X[:int(m*0.6),:]
Y_train = Y[:int(m*0.6)]
X_dev = X[int(m*0.6):int(m*0.8),:]
Y_dev = Y[int(m*0.6):int(m*0.8)]
X_test = X[int(m*0.6):,:]
Y_test = Y[int(m*0.6):]
```

```python
#small plotting example
fig = plt.figure(figsize=(17,5))

ax1 = fig.add_subplot(1,3,1) #rows,coloumns,index
ax1.plot(X[:,0],Y,'o')
ax1.set_xlabel('Feature1')
ax1.set_ylabel('Output')
```

```python
# applying linear regression
lr = LinearRegression()
lr.fit(X_train,Y_train)
print(lr.intercept_)
print(lr.coef_)
approcFunc = lambda x: lr.intercept_ + np.dot(x,lr.coef_)

#prediction function outpu:
plt.plot(X_train[:,0],Y_train,'o')
plt.plot(X_train[:,0],lr.predict(X_train),'o')
plt.show()
```

```python
# evaluation of preformance
def RMSE(output,prediction):
    return np.sqrt(np.mean((output-prediction)**2))

prediction = lr.predict(X_dev)
print(RMSE(Y_dev, prediction))
```

```python
# scaling the data features
scaler = preprocessing.StandardScaler() # class for scaling in python
X_train_scaled = scaler.fit_transform(X_train)
# mean and variance used for scaling the X_train data (fit, then transform)
print(scaler.mean_)
print(scaler.var_)
#now we transform the dev data
X_dev_scaled = scaler.transform(X_dev)
```

```python
# fitting with scaled data
lr = LinearRegression()
lr.fit(X_train_scaled, Y_train)
predictions = lr.predict(X_dev_scaled)
dev_error = RMSE(Y_dev, predictions)
print(dev_error)
```

```python
# In case of non-linear features
poly = preprocessing.PolynomialFeatures(degree=2, include_bias=False)
# makes more colomns of data with the specidied polynomial degree
# using a loop :
error_vec_train = []
error_vec_dev = []
for i in range(1,7):
    poly = preprocessing.PolynomialFeatures(degree=i, include_bias=False)
    X_train_scaled_d2 = poly.fit_transform(X_train_scaled)
    X_dev_scaled_d2 = poly.fit_transform(X_dev_scaled)
    
    lr = LinearRegression(fit_intercept=True)
    lr.fit(X_train_scaled_d2, Y_train)
    predictions = lr.predict(X_dev_scaled_d2)
    error_dev = RMSE(Y_dev, predictions)
    error_train = RMSE(Y_train, lr.predict(X_train_scaled_d2))
    print(error_dev)
    error_vec_train.append(error_train)
    error_vec_dev.append(error_dev)
plt.plot(range(1,7), error_vec_train, label='train')
plt.plot(range(1,7), error_vec_dev, label='dev')
plt.xlabel("Degrees")
plt.ylabel("RMSE")
plt.legend()
plt.show()
    
```
